```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('wine_data.csv')
print(df)
```

```python
df.isna().sum()
```

 Frequency

```python
#We find the mode of the quality for the most frequntly occuring wine quality

print("The frequently occuring quality is:",df['quality'].mode())
```

```python
#Highest and lowest in quality column
M_A_X=df['quality'].max()
M_I_N=df['quality'].min()
print("The highest quality is:",M_A_X)
print("The lowest quality is:",M_I_N)
```

2) Correlation

```python
#Fixed acidity correlation to quality
corr_fixed_acidity=df['fixed acidity'].corr(df['quality'])
print('The correlated value is:',corr_fixed_acidity)
```

```python
#We will find the correlation between alcohol and quality to find how does it affect wine quality
corr_alcohol=df['alcohol'].corr(df['quality'])
print('The correlated value is:',corr_alcohol)
```

```python
'''This indicates a moderate positive correlation.
This means that generally, as the alcohol content of the wine increases, the quality of the wine also tends to increase'''
```

```python
#Sulphur content correlated to quality
corr_sulphur_content=df['free sulfur dioxide'].corr(df['quality'])
print('The correlated value is:',corr_sulphur_content)
```

3) Avg residual sugar

```python
avg_residual_max=df[df['quality']==df['quality'].max()]['residual sugar'].mean
avg_residual_min=df[df['quality']==df['quality'].min()]['residual sugar'].mean
print('The max avg residual sugar is:',avg_residual_max)
print('The min avg residual sugar is:',avg_residual_min)
```

4) Volatile acidity

```python
corr_value=df['volatile acidity'].corr(df['quality'])
print('The correlated value or the effect is:',corr_value)
```

```python
''''Yes, volatile acidity does have an effect on the quality of wine samples.
The correlation between volatile acidity and quality is approximately -0.39.
This indicates a moderate negative correlation.
This means that as volatile acidity increases, the quality of the wine tends to decrease'''
```

5) Random Forests and Decision Tree

```python
X=df.drop('quality',axis=1)
y=df['quality']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
print(X_train)
print(y_train)
print(X_test)
print(y_test)
```

```python
#Preprocess the data
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
print(X_train)
print(X_test)
```

```python
#Decision Tree
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
#Train the model
classifier.fit(X_train, y_train)
```

```python
#Prediction
y_test=classifier.predict(X_test)
y_pred = classifier.predict(X_test)
print(y_pred)
```

```python
#Accuracy scores
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print("Decision Tree Model Accuracy:", accuracy)
```

```python
#Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)
```

```python
#prediction
y_RF_pred=classifier.predict(X_test)
print(y_RF_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_RF_pred)
print("Random Forests model accuracy is:", accuracy)
```

```python
'''Conclusion
Decision Tree model appears to be better because it achieved a perfect accuracy score of 1.0,
while the Random Forest model achieved a lower score of 0.63125'''
```
